```python
#importing the required libraries for the scrapping
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#Extracting the dataset from the website

url = (
    "https://www.imdb.com/title/tt10366206/reviews/_ajax?ref_=undefined&paginationKey={}"
)
key = ""
john_wick_data = {"title": [], "review": [], "rating": [], "date":[]}

while True:
    response = requests.get(url.format(key))
    soup = BeautifulSoup(response.content, "html.parser")
    # Find the pagination key
    pagination_key = soup.find("div", class_="load-more-data")
    if not pagination_key:
        break

    # Update the `key` variable in-order to scrape more reviews
    key = pagination_key["data-key"]
    for title, review, rating, date in zip(
        soup.find_all(class_="title"), soup.find_all(class_="text show-more__control"), soup.find_all( class_ = "rating-other-user-rating"),
        soup.find_all(class_ = "review-date")
    ):
        john_wick_data["title"].append(title.get_text(strip=True))
        john_wick_data["review"].append(review.get_text())
        john_wick_data["rating"].append(rating.get_text(strip=True))
        john_wick_data["date"].append(date.get_text())
```

```python
#putting the data in a dataframe
df = pd.DataFrame(john_wick_data)
df
```

```python
#Saving to data as csv file
df.to_csv('John_Wick_4')
```
